```python
# =========================================
# TEMA:    Visualización de Canales RGB
# CURSO:   Procesamiento de Imágenes y Visión Computacional
# EJEMPLO: Mostrar imagen original y separación en canales Rojo, Verde y Azul
# =========================================

# Importación de librerias necesarias
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Paso 1: Cargar la imagen en formato BGR (como lo hace OpenCV por defecto)
imagen_bgr = cv2.imread('prueba.jpeg') # Nota: Reemplaza 'prueba.jpeg' con el nombre del archivo que estés usando

# Convertir de BGR a RGB para que se vea correctamente con matplotlib
imagen_rgb = cv2.cvtColor(imagen_bgr, cv2.COLOR_BGR2RGB)

# Paso 2: Separar los 3 canales de color (Rojo, Verde, Azul)
canal_rojo  = imagen_rgb[:, :, 0]  # Canal Rojo
canal_verde = imagen_rgb[:, :, 1]  # Canal Verde
canal_azul  = imagen_rgb[:, :, 2]  # Canal Azul

# Paso 3: Mostrar la imagen original y los canales por separado
plt.figure(figsize=(12, 4))

# Imagen original
plt.subplot(1, 4, 1)
plt.imshow(imagen_rgb)
plt.title("Imagen Original (RGB)")
plt.axis('off')

# Canal Rojo (visualizado con cmap='Reds')
plt.subplot(1, 4, 2)
plt.imshow(canal_rojo, cmap='Reds')
plt.title("Canal Rojo")
plt.axis('off')

# Canal Verde (visualizado con cmap='Greens')
plt.subplot(1, 4, 3)
plt.imshow(canal_verde, cmap='Greens')
plt.title("Canal Verde")
plt.axis('off')

# Canal Azul (visualizado con cmap='Blues')
plt.subplot(1, 4, 4)
plt.imshow(canal_azul, cmap='Blues')
plt.title("Canal Azul")
plt.axis('off')

plt.tight_layout()
plt.show()

# ============================================================
# ACTIVIDAD PARA EL PARTICIPANTE:
# 1. Combina solo dos canales, por ejemplo Rojo + Verde (deja el Azul en 0)
# 2. Prueba también Verde + Azul y Rojo + Azul
# 3. Finalmente, vuelve a combinar los tres canales para reconstruir la imagen RGB original
#    Sugerencia, usa: np.stack([canal_rojo, canal_verde, canal_azul], axis=2)
# ============================================================
```
