```python
# =========================================
# TEMA:    Visualización de Canales RGB y HSV
# CURSO:   Procesamiento de Imágenes y Visión Computacional
# EJEMPLO: Mostrar imagen original y separación de canales RGB y HSV
# =========================================

# Importación de librerías necesarias
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Paso 1: Cargar la imagen en formato BGR (por defecto en OpenCV)
imagen_bgr = cv2.imread('prueba.jpeg')  # Nota: Reemplaza 'prueba.jpeg' con tu archivo

# Validar que la imagen se haya cargado correctamente
if imagen_bgr is None:
    raise FileNotFoundError("La imagen no se encontró. Verifica la ruta del archivo.")

# Convertir de BGR a RGB
imagen_rgb = cv2.cvtColor(imagen_bgr, cv2.COLOR_BGR2RGB)

# Convertir de RGB a HSV
imagen_hsv = cv2.cvtColor(imagen_rgb, cv2.COLOR_RGB2HSV)

# Separar canales RGB
canal_rojo = imagen_rgb[:, :, 0]
canal_verde = imagen_rgb[:, :, 1]
canal_azul = imagen_rgb[:, :, 2]

# Separar canales HSV
canal_h = imagen_hsv[:, :, 0]
canal_s = imagen_hsv[:, :, 1]
canal_v = imagen_hsv[:, :, 2]

# Paso 2: Mostrar imagen original y canales separados
fig, axs = plt.subplots(2, 4, figsize=(15, 7))

# Imagen original RGB
axs[0, 0].imshow(imagen_rgb)
axs[0, 0].set_title('Imagen RGB')
axs[0, 0].axis('off')

# Canales RGB con cmap adecuados
axs[0, 1].imshow(canal_rojo, cmap='Reds')
axs[0, 1].set_title('Canal Rojo')
axs[0, 1].axis('off')

axs[0, 2].imshow(canal_verde, cmap='Greens')
axs[0, 2].set_title('Canal Verde')
axs[0, 2].axis('off')

axs[0, 3].imshow(canal_azul, cmap='Blues')
axs[0, 3].set_title('Canal Azul')
axs[0, 3].axis('off')

# Imagen HSV mostrada directamente (incorrecto)
axs[1, 0].imshow(imagen_hsv)
axs[1, 0].set_title('Imagen HSV (sin convertir)')
axs[1, 0].axis('off')

# Canales HSV
axs[1, 1].imshow(canal_h, cmap='hsv')
axs[1, 1].set_title('Matiz (Hue)')
axs[1, 1].axis('off')

axs[1, 2].imshow(canal_s, cmap='gray')
axs[1, 2].set_title('Saturación')
axs[1, 2].axis('off')

axs[1, 3].imshow(canal_v, cmap='gray')
axs[1, 3].set_title('Valor (Brightness)')
axs[1, 3].axis('off')

plt.tight_layout()
plt.show()

# ================================================
# IMPORTANTE:
# ⚠️ La celda "Imagen HSV (sin convertir)" mostrará colores distorsionados.
# Esto es intencional para que el estudiante observe qué ocurre cuando se
# visualiza una imagen HSV directamente con imshow().
# ✅ En la práctica, siempre debe convertirse primero a RGB:
#     imagen_hsv_rgb = cv2.cvtColor(imagen_hsv, cv2.COLOR_HSV2RGB)
# ================================================

# ============================================================
# ACTIVIDAD PARA EL PARTICIPANTE:
# 1. Observa la diferencia visual entre la imagen RGB y la HSV directa.
# 2. Explica por qué los colores se ven distorsionados en HSV sin conversión.
# 3. Aplica luego la conversión a RGB y vuelve a visualizar la imagen HSV correctamente.
# ============================================================
```
